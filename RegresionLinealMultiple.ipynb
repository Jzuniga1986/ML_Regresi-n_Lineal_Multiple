<h4 Style='color:#ecbd5c'>Preparando el modelo de regresion lineal<h4>
    
<p Style='text-align: justify; font-style: normal; font-weight: normal'> Una vez transitada la etapa de analisis de datos y seleccion de las variables a utilizar en el modelo de regresion lineal se procede con la creacion de los set de datos de entrenamiento y pruebas.
<br>
<br>    
Train/Test Split implica dividir el conjunto de datos en conjuntos de entrenamiento y prueba que son mutuamente excluyentes. Despu칠s de lo cual, entrena con el conjunto de entrenamiento y prueba con el conjunto de prueba. Esto proporcionar치 una evaluaci칩n m치s precisa de la precisi칩n fuera de la muestra porque el conjunto de datos de prueba no forma parte del conjunto de datos que se utiliz칩 para entrenar el modelo. Por lo tanto, nos da una mejor comprensi칩n de qu칠 tan bien generaliza nuestro modelo en nuevos datos.
<br>
<br> 
Esto significa que conocemos el resultado de cada punto de datos en el conjunto de datos de prueba, 춰lo que lo hace genial para probar! Dado que estos datos no se han utilizado para entrenar el modelo, el modelo no tiene conocimiento del resultado de estos puntos de datos. Entonces, en esencia, es realmente una prueba fuera de muestra.
<br>
<br> 
Dividamos nuestro conjunto de datos en conjuntos de entrenamiento y prueba. El 80 % de todo el conjunto de datos se utilizar치 para entrenamiento y el 20 % para pruebas. Creamos una m치scara para seleccionar filas aleatorias usando la funci칩n np.random.rand():
</p>
<br>
<br>

```python
msk = np.random.rand(len(df)) < 0.8 #Esto selecciona aleatoriamente el 80% de las observaciones y la amacena en msk
train = cdf[msk] #aca se indica que el set de entrenamiento esta conformado por el 80% de las observaciones
test = cdf[~msk] #aca se especifica que el set de prueba esta conformado por el 20% de la data restante.
```

<h4 Style='color:#ecbd5c'>Generacion del modelo de Regresion Lineal<h4>
    
<p Style='text-align: justify; font-style: normal; font-weight: normal'> En realidad, existen m칰ltiples variables que impactan la co2emisi칩n. Cuando est치 presente m치s de una variable independiente, el proceso se denomina regresi칩n lineal m칰ltiple. Un ejemplo de regresi칩n lineal m칰ltiple es la predicci칩n de co2emisiones utilizando las caracter칤sticas FUELCONSUMPTION_COMB, EngineSize y Cylinders de carros. Lo bueno aqu칤 es que el modelo de regresi칩n lineal m칰ltiple es la extensi칩n del modelo de regresi칩n lineal simple.
</p>

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

<p>Como se ha mencionado los coeficientes e interseccion son los parametros de la linea ajustada, sin embargo, al ser un modelo de regresi칩n lineal m칰litple esos parametros podr칤an presentar alguna inconsistencia, para solucionar esto podemo optar por utilizar el metodo simple de minimos cuadrados que esta en la librer칤a scikit learn.
<br>
<br>
M칤nimos cuadrados ordinarios (OLS)
<br>
<br>
OLS es un m칠todo para estimar los par치metros desconocidos en un modelo de regresi칩n lineal. OLS elige los par치metros de una funci칩n lineal de un conjunto de variables explicativas minimizando la suma de los cuadrados de las diferencias entre la variable dependiente objetivo y las predichas por la funci칩n lineal. En otras palabras, intenta minimizar la suma de los errores cuadr치ticos (SSE) o el error cuadr치tico medio (MSE) entre la variable objetivo (y) y nuestra salida pronosticada ( 洧녽팩 ) sobre todas las muestras en el conjunto de datos.
<br>
<br>
OLS puede encontrar los mejores par치metros utilizando los siguientes m칠todos:
<br>
<br>
Resolviendo los par치metros del modelo anal칤ticamente usando ecuaciones de forma cerrada
Utilizando un algoritmo de optimizaci칩n (Gradient Descent, Stochastic Gradient Descent, M칠todo de Newton, etc.)
<br>
<br>
Aclarado esto, procedemos a realizar la predicci칩n de los valores
</p>

```python
y_hat= regr.predict(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
x = np.asanyarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(test[['CO2EMISSIONS']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

print('Variance score: %.2f' % regr.score(x, y))
```

<p Style='text-align: justify; font-style: normal; font-weight: normal'> Como se puede observar el resultado del modelo genera una confianza del 86%, de manera que este resulta ser m치s efectivo que la regresi칩n lineal simple que gener칩 77%. Por lo tanto al comparar los 2 modelos brinda un mejor resultado la regresi칩n lienal multiple</p>
